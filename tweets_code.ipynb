```python
x = df1.iloc[:,2].values
```

```python
x
```

```python
pos=0
neg=0
neu=0
y=0
X = []

for tweet in x:
    y=y+1
    #tweet=processTweet(tweet)
    analysis = TextBlob(tweet) 

    if analysis.sentiment.polarity >0:
     pos=pos+1
     t = 'Positive'
     X.append(t)

    elif analysis.sentiment.polarity == 0:
      neu=neu+1
      t = 'Neutral'
      X.append(t)

    else:
       neg=neg+1
       t = 'Negative'
       X.append(t)
```

```python
print(pos)
print(neg)
print(neu)
```

```python
X
```

```python
df1['overall_polarity'] = X
```
